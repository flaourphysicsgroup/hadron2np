## 2. 生成一个衰变过程对象，并查看其属性

```python
process = hadron2np.new_decay_process(['B+', 'K+', 'phi', 'phi'], basis='DLEFT(S/P)', ff_imp='LCSR-pole 2004')
```

```python
process.index
```

```python
wc_here = np.zeros((3, 3, 2, 2), dtype=complex)
random_num = np.linspace(0.1, 1, 10)
for ii in random_num:
    wc_here[*process.index] = ii
    print(process.width(wcs={'L_S_dphi2': wc_here}, m_dm=[1, 1]))
```
